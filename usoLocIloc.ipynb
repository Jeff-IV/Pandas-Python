# Manejo de propiedades loc, iloc, et y at


## Cargar el dataframe


```python
import pandas as pd

data = "dataset/Pokemon.csv"

df = pd.read_csv(data)

df.set_index("Name", inplace = True)
df.head()
```

## Selecci√≥n de un grupo de filas

```python
df.iloc[2:8] #iloc √∫nicamente recibe enteros
```

```python
df.loc[[
    "Charmeleon", "Pikachu", "Gengar"
]] 
```

## Selecci√≥n de un grupo de filas y columnas


```python
df.iloc[2:8, [0,4,2,3]]
```

```python
filas = ["Charizard", "Raichu", "Pidgey", "Mewtwo"]

columnas = ["HP", "Generation","Legendary"]

df.loc[filas, columnas]
```

## Selecci√≥n de un registro en especifico


```python
df.iloc[0]
```

```python
df.loc["Onix"]
```

## Selecci√≥n de una celda en especifico por posici√≥n 

```python
df.iat[0, 1]
```

## Selecci√≥n de una celda en part√≠cular usando etiquetas

```python
df.at["Pikachu", "Type 1"]
```
