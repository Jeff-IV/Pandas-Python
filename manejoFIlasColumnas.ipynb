# Manejo de columnas


## Carguemos el dataframe

```python
import pandas as pd

data = "dataset/Pokemon.csv"

df = pd.read_csv(data)

#Visualizar las coolumnas 
df.columns
```

## Establezcamos la columna Name como indice del dataframe

```python
df.set_index("Name", inplace = True) #inplace = True para hacer el cambio en el df original

df.head()
```

## Obtención de un subconjunto de columnas 

```python
df[["Type 1", "Type 2", "Generation"]].head()
```

## Obtención de un subconjunto de columnas usando filter

```python
df.filter(items=["Type 1", "Type 2", "Generation"]).head()
```

## Obtención de un subconjunto de columnas usando regular expression

```python
df.filter(regex="e$", axis=1).head() # regex="e$" para referirnos a todos los que terminen con e, axis=1 para que use la expresión regular en las columnas 
```

```python
df.filter(like="pe", axis=1).head() # like="pe" para referirnos a todos los que contengan pe  
```

## Ordenar columnas en el dataframe

```python
df.sort_index(axis=1, ascending=True).head()
```

## Cambiar nombres a columnas

```python
df.rename(columns = {
    "Legendary" : "isLegendary",
    "HP" : "Health"
}).head()
```

# Manejo de filas

Se basa en el indice para la obtención de las filas (ya sea el númerico (posición del elemento), o el dado por nosotros)

## Obtención de un subconjunto de registros


```python
df[3:6] 
```

```python
df.loc[["Charmander", "Pikachu", "Pidgey", "Gengar", "Squirtle"]]
```

## Ordenamiento del indice

```python
df.sort_index(axis=0, ascending=False).head()
```

## Obtención de un subconjunto de filas usando regular expression


```python
df.filter(regex="u$", axis=0).head() #u$ para registros que terminen en u 
```

```python
df.filter(like="cha", axis=0).head() #registros que contengan cha
```

## Cambiar nombre de filas

```python
df.rename( index={
    "Charmander" : "Dani",
    "Charmeleon" : "Dani Evolucionada"
}).head(6)
```
